## 3. ***Data Wrangling***

### Data Wrangling Code

````python
# Write your code to make your dataset analysis ready.


df.drop_duplicates().reset_index(inplace=True)  # In case there is some duplicate values

# Fix data types
df["Age"] = df["Age"].astype(int)
df["SSN"] = df["SSN"].astype(str)

# Remove invalid ages
df = df[(df["Age"] >= 18) & (df["Age"] <= 80)]

# Standardize text columns
text_cols = ["Occupation", "Credit_Mix", "Payment_Behaviour", "Payment_of_Min_Amount"]
for i in text_cols:
    df[i] = df[i].astype(str).str.strip().str.lower()

# 3. Handle Missing & Invalid Numeric Values

numeric_cols = [
    "Num_Bank_Accounts",
    "Num_Credit_Card", "Num_of_Loan", "Delay_from_due_date",
    "Num_of_Delayed_Payment",
    "Num_Credit_Inquiries", "Total_EMI_per_month",
    "Amount_invested_monthly"
]

for col in numeric_cols:
  df[col] = df[col].fillna(df[col].median())

# Debt to income ratio
df["debt_to_income_ratio"] = df["Outstanding_Debt"] / (df["Annual_Income"] + 1)

# Risk flags
df["high_utilization_flag"] = np.where(df["Credit_Utilization_Ratio"] > 0.7, 1, 0)
df["high_delay_flag"] = np.where(df["Num_of_Delayed_Payment"] > 5, 1, 0)
df["multi_account_flag"] = np.where(df["Num_Bank_Accounts"] > 5, 1, 0)
df["salary_mismatch_flag"] = np.where(
    df["Monthly_Inhand_Salary"] * 12 > df["Annual_Income"] * 1.2, 1, 0
)
# 5. Target Variable Creation (Fraud Flag)

df["Fraud_Flag"] = np.where(
    (df["Credit_Score"].str.lower() == "poor") |
    (df["high_utilization_flag"] == 1) |
    (df["high_delay_flag"] == 1) |
    (df["debt_to_income_ratio"] > 1),
    1, 0
)

# 6. Remove Leakage & Irrelevant Columns (VERY IMPORTANT)

cols_to_drop = ["ID", "SSN", "Credit_Score"]
df = df.drop(columns=cols_to_drop)

# 7. Final Dataset Check (EDA Ready)

print("Final Shape:", df.shape)
print("Fraud Rate:", round(df["Fraud_Flag"].mean(), 3))
print(df.head())







````

### What all manipulations have you done and insights you found?

###  Data Cleaning

- Converted incorrect data types (Age, SSN, numerical fields) to appropriate formats.
- Removed invalid age records (<18 and >80) to maintain realistic customer profiles.
- Removed duplicate records to avoid repeated analysis.
- Handled missing and infinite values using median imputation to preserve data distribution.
- Standardized categorical values (occupation, credit mix, payment behavior) for consistency.

### Removal of Irrelevant & Leakage Columns

- Removed identifiers such as **Name, SSN, and ID** as they add no analytical value.
- Removed **Credit_Score** after using it to define the fraud flag to prevent data leakage.
- Retained only behavioral and financial features relevant for fraud detection.

###  Feature Engineering (Fraud Signals)

- Created **debt-to-income ratio** to capture over-leverage risk.
- Added **high credit utilization flag** to detect overspending behavior.
- Added **high delayed payment flag** to identify repeat defaulters.
- Added **multiple bank account flag** to capture potential mule accounts.
- Added **salary mismatch flag** to identify possible fake income declarations.
- Created **rule-based fraud flag** to classify high-risk customers.

###  Data Structuring

- Maintained data at **monthly customer level** to capture behavioral patterns.
- Prepared final dataset with **clean, consistent, and business-relevant features**.
- Saved cleaned dataset for EDA, dashboarding, and reporting.
